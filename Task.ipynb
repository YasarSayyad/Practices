```python
from datetime import datetime

class MutualFundInvestment:
    def __init__(self, scheme_code, start_date, end_date, initial_investment, final_investment):
        self.scheme_code = scheme_code
        self.start_date = datetime.strptime(start_date, "%d-%m-%Y")
        self.end_date = datetime.strptime(end_date, "%d-%m-%Y")
        self.initial_investment = initial_investment
        self.final_investment = final_investment

    def calculate_profit(self):
        # Calculate the holding period in years
        holding_period_years = (self.end_date - self.start_date).days / 365.25

        # Calculate the compound annual growth rate (CAGR)
        cagr = (self.final_investment / self.initial_investment) ** (1 / holding_period_years) - 1

        # Calculate the profit
        profit = self.initial_investment * (1 + cagr) ** holding_period_years - self.initial_investment

        return profit

# Example usage:
scheme_code = "101206"
start_date = "26-07-2023"
end_date = "18-10-2023"
initial_investment = 1000000
final_investment = 1500000

investment = MutualFundInvestment(scheme_code, start_date, end_date, initial_investment, final_investment)
profit = investment.calculate_profit()

print(f"Profit for Scheme {scheme_code} from {start_date} to {end_date}: ${profit:.2f}")
```

```python
# Get the Net Asset Value (NAV) of the mutual fund on both the purchase date (26-07-2023) and redemption date (18-10-2023) using the API: https://api.mfapi.in/mf/101206.
import requests

def get_nav(api_url, date):
    response = requests.get(f"{api_url}/{date}")
    
    if response.status_code == 200:
        nav_data = response.json()
        return nav_data['data']['nav']
    else:
        print(f"Error fetching NAV data. Status code: {response.status_code}")
        return None

# Example usage:
api_url = "https://api.mfapi.in/mf/101206"
purchase_date = "26-07-2023"
redemption_date = "18-10-2023"

nav_purchase_date = get_nav(api_url, purchase_date)
nav_redemption_date = get_nav(api_url, redemption_date)

if nav_purchase_date is not None and nav_redemption_date is not None:
    print(f"NAV on {purchase_date}: {nav_purchase_date:.5f}")
    print(f"NAV on {redemption_date}: {nav_redemption_date:.5f}")
```

```python
# Calculate the net profit:
from datetime import datetime

class MutualFundInvestment:
    def __init__(self, scheme_code, purchase_date, start_date, end_date, initial_investment, final_investment):
        self.scheme_code = scheme_code
        self.purchase_date = datetime.strptime(purchase_date, "%Y-%m-%d")
        self.start_date = datetime.strptime(start_date, "%Y-%m-%d")
        self.end_date = datetime.strptime(end_date, "%Y-%m-%d")
        self.initial_investment = initial_investment
        self.final_investment = final_investment

    def calculate_units_allotted(self, nav_on_purchase_date):
        # Calculate the number of units allotted on the purchase date
        units_allotted = self.initial_investment / nav_on_purchase_date
        return units_allotted

    def calculate_units_value(self, units, nav_on_redemption_date):
        # Calculate the value of units on the redemption date
        units_value = units * nav_on_redemption_date
        return units_value

    def calculate_net_profit(self, units_value_on_redemption):
        # Calculate the net profit
        net_profit = units_value_on_redemption - self.initial_investment
        return net_profit

    def calculate_profit(self):
        # Calculate the holding period in years
        holding_period_years = (self.end_date - self.start_date).days / 365.25

        # Calculate the compound annual growth rate (CAGR)
        cagr = (self.final_investment / self.initial_investment) ** (1 / holding_period_years) - 1

        # Calculate the profit
        profit = self.initial_investment * (1 + cagr) ** holding_period_years - self.initial_investment

        return profit

# Example usage:
scheme_code = "XYZ123"
purchase_date = "2023-07-26"
start_date = "2020-01-01"
end_date = "2023-01-01"
initial_investment = 1000000
final_investment = 1500000
nav_on_purchase_date = 3682.28990
nav_on_redemption_date = 3737.30240

investment = MutualFundInvestment(scheme_code, purchase_date, start_date, end_date, initial_investment, final_investment)
units_allotted = investment.calculate_units_allotted(nav_on_purchase_date)
units_value_on_redemption = investment.calculate_units_value(units_allotted, nav_on_redemption_date)
net_profit = investment.calculate_net_profit(units_value_on_redemption)
profit = investment.calculate_profit()

print(f"Number of units allotted on {purchase_date}: {units_allotted:.2f} units")
print(f"Value of {units_allotted:.2f} units on {redemption_date}: ${units_value_on_redemption:.2f}")
print(f"Net Profit: ${net_profit:.2f}")
print(f"Profit for Scheme {scheme_code} from {start_date} to {end_date}: ${profit:.2f}")
```
